```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Statistics for YouCook2

```python
df = pd.read_csv('youcook2_withtime.csv')

df.head()
```

```python
grouped_data = df.groupby('year').size()

print(grouped_data)
```

```python
# Plotting
plt.figure(figsize=(10, 6))
bars = plt.bar(grouped_data.index, grouped_data.values, color='skyblue')

# Adding labels to each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval, int(yval), va='bottom', ha='center', fontsize=12)

plt.title('YouCook2 Number Of Videos vs Yearr')
plt.xlabel('Year')
plt.ylabel('Number of Videos')
plt.xticks(grouped_data.index)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
