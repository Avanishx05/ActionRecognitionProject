```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plot_time_distribution(df, dt_name):
    grouped_data = df.groupby('year').size()
    # Plotting
    plt.figure(figsize=(10, 6))
    bars = plt.bar(grouped_data.index, grouped_data.values, color='skyblue')

    # Adding labels to each bar
    for bar in bars:
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2.0, yval, int(yval), va='bottom', ha='center', fontsize=12)

    plt.title(f'{dt_name} Number Of Videos vs Year')
    plt.xlabel('Year')
    plt.ylabel('Number of Videos')
    plt.xticks(grouped_data.index)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
```

### Statistics for YouCook2

#### Train-Val Data

```python
df_tv = pd.read_csv('DatasetsWithTime/youcook2trainval_withtime.csv')

df_tv.head()
```

```python
plot_time_distribution(df_tv, "YouCook2 TrainVal")
```

#### Test Data

```python
df_t = pd.read_csv('DatasetsWithTime/youcook2test_withtime.csv')

df_t.head()
```

```python
plot_time_distribution(df_t, "YouCook2 Test")
```

#### Total Data

```python
df_yc2total = pd.concat([df_tv, df_t], axis=0, ignore_index=True)

plot_time_distribution(df_yc2total, "YouCook2 Total")
```
